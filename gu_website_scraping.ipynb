```python
def scrapeAList(list_of_urls):
    '''
    Takes a list of urls as the input parameter, 
    returns the text contents scraped directly from these urls, 
    and a list of sub-urls linked from the given page.
    '''
    content = []
    pattern = '^https:\/\/'
    for url in list_of_urls:
        if bool(re.search(pattern,url)) == True:
            response = requests.get(url)
            soup = BeautifulSoup(response.text, 'html.parser')
            content.append(soup.find_all('p'))
            suburl = []
            for link in soup.find_all('a'):
                suburl.append(link.get('href'))
        else:
            pass
    return(content,suburl)

#scrapeAList(['https://www.georgetown.edu/','#navigation'])
```

```python
URL_init = 'https://www.georgetown.edu/'

def scrapeItAll(URL_init):
    '''
    Given the initial gu homepage, 
    scrape three layers of sub-urls
    '''
    layers = 0
    current_urllist = [URL_init]
    all_content = []
    all_subs = []
    all_subs.append(current_urllist)
    while layers < 3:
        content, suburls = scrapeAList(current_urllist)
        all_content.append(content)
        all_subs.append(suburls)
        # update layers and current_urllst 
        layers = layers + 1
        current_urllist = suburls
    return(all_content, all_subs)
    
           
texts, hrefs = scrapeItAll(URL_init)

    
```

```python
with open('contents.txt', 'w') as f:
    for lst in texts:
        f.write("\n\n")
        for string in lst:
            f.write(str(string) + '\n')
```

```python
with open('hrefs.txt', 'w') as f:
    for url in hrefs:
        f.write("\n\n")
        for item in url:
            f.write(str(item) + '\n')
```
